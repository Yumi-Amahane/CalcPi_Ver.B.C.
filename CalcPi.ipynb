内側多角形の辺として斜辺の1辺の大きさ$c_{in}$を求める  
$c_{in}^2=a^2+b^2-2ab\cos C$  
$C=2\pi/N$ラジアン(上図では$N=4$)

```python
plt.figure(figsize=(3,3))
plt.plot(X,Y,color="b")
plt.plot(X,np.array(Y)*-1,color="b")

plt.plot(np.arange(-1,0.1,0.1),np.arange(1,-0.1,-0.1),color="r")
plt.plot(np.arange(-1,1.1,0.1),[1]*21,color="r")
plt.plot(np.arange(0,1.1,0.1),np.arange(0,1.1,0.1),color="r")

plt.plot([0,0],[0,1],linestyle="dashed",color="g")

plt.plot([-1]*21,np.arange(1,-1.1,-0.1),color="r",linestyle="dashed")
plt.plot([1]*21,np.arange(-1,1.1,0.1),color="r",linestyle="dashed")
plt.plot(np.arange(-1,1.1,0.1),[-1]*21,color="r",linestyle="dashed")

plt.xlabel("x")
plt.ylabel("y")
plt.savefig("4D")
```

外側多角形の1辺の長さ$c_{out}$として斜辺の大きさを求める  
$c_{out}=2\tan \frac C 2$  
$C=2\pi/N$(上図では$N=4$)

$c_{in}$と$c_{out}$ともに$N$倍して多角形の辺の長さの総和にする

$N=4$のとき、円周長さ$R$は  
$4\sqrt 2\le R \le 2\times 4$

直径が2より、直径と円周率の関係（円周率の定義）より円周率を求められる

$N=4$のとき、円周率は  
$2\sqrt 2\le R \le 4$

$N$を大きくして精度を上げる

$N$を大きくした際の$\cos$、$\tan$に関しては計算機が無くとも倍角の公式  
$\sin(\theta)=2\sin(\frac \theta 2)\cos(\frac \theta 2)$  
$\cos(\theta)=\cos^2(\frac \theta 2)-\sin^2(\frac \theta 2)$  
をつかえばいいね。$N$をだんだん倍々にして行けるね

九十六角形の時

```python
N=96
ArgC=2*math.pi/N
c_in=(2-2*math.cos(ArgC))**0.5*N
c_out=(2*math.tan(ArgC/2)*N)
```

```python
c_in=[]
c_out=[]

Itr=np.arange(4,50)
for N in Itr:
    ArgC=2*math.pi/N
    c_in.append((2-2*math.cos(ArgC))**0.5*N)
    c_out.append(2*math.tan(ArgC/2)*N)
```

```python
c_in=[]
c_out=[]

Nmin=1000
Itr=np.arange(Nmin,Nmin*10)
for N in Itr:
    ArgC=2*math.pi/N
    c_in.append((2-2*math.cos(ArgC))**0.5*N)
    c_out.append(2*math.tan(ArgC/2)*N)
    
    
plt.plot(Itr,np.array(c_in)/2)
plt.plot(Itr,np.array(c_out)/2)
plt.legend(["$c_{in}$","$c_{out}$"])

plt.xlabel("N")
plt.ylabel("$\pi$")

plt.hlines(math.pi,Itr[0],Itr[-1],linestyle="dashed")
plt.xscale("log")
plt.savefig("A2")
```
